Estadística
===========

Para el aprendizaje automatizado, la estadística es una herramienta fundamental para entender los datos con los que se trabaja. En este capítulo se presentan los conceptos básicos de estadística que se utilizan en el aprendizaje automatizado.

De interés para el aprendizaje automatizado son los conceptos de:

- **Media**: Es el promedio de los datos. Se calcula sumando todos los datos y dividiendo entre el número de datos. Es una medida de la tendencia de los datos.
- **Moda**: Es el dato que más se repite en los datos. Nos da otra idea de la tendencia de los datos.
- **Desvío**: Es la diferencia entre cada dato y la media. Es una medida de la dispersión de los datos.
- **Varianza**: Es el promedio de los desvíos al cuadrado. 
- **Mínimo y máximo**: El dato más pequeño y el dato más grande. Establecen el rango de los datos.
- **Datos atípicos**: Son datos que se alejan mucho de la media. Pueden ser errores de medición o datos que no siguen el patrón de los demás datos. Introducen ruido en los datos.
- **Distribución**: Es la forma en que se distribuyen los datos. Puede ser normal, uniforme, etc.

Estos datos son importantes a la hora de seleccionar el algoritmo a utilizar y para entender los resultados obtenidos. Por ejemplo, varios algoritmos de aprendizaje automatizado asumen que los datos tienen una distribución normal, o que la escala de estos datos es la misma, lo cual puede requerir una normalización de los datos en caso de que estas suposiciones no se cumplan.

## Python

En Python, la librería `numpy` provee funciones para calcular la media, la moda, el desvío, la varianza, el mínimo y el máximo de un conjunto de datos. Al mismo tiempo, la librería `matplotlib` provee funciones para graficar los datos y ver su distribución.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

datos_1 = np.random.binomial(45, 0.3, 1000)
datos_2 = np.append(np.random.binomial(20, 0.2, 750), [np.nan] * 250)
datos_3 = np.append(np.random.binomial(78, 0.4, 800), [np.nan] * 200)

data_df = pd.DataFrame({'datos_1': datos_1, 'datos_2': datos_2, 'datos_3': datos_3})

statistic = pd.DataFrame([], index=['mean', 'std', 'min', 'max', 'median', 'mising'])
for column in statistic.columns:
    statistic[column] = statistic[column].astype('float')

for column in data_df.columns:
    if data_df[column].dtype in ('object', 'category', 'string'):
        statistic[column] = [np.nan] * 5 + [data_df[column].isna().sum()]
        continue

    if data_df[column].isna().all():
        statistic[column] = [np.nan] * 5 + [len(data_df[column])]
        continue

    statistic[column] = [
        data_df[column].mean(),
        data_df[column].std(),
        data_df[column].min(),
        data_df[column].max(),
        data_df[column].median(),
        np.isnan(data_df[column]).sum()
    ]
statistic = statistic.T
print('Estadísticas de los datos')
print(statistic)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].hist(data_df['datos_1'], bins=10, color='red')
ax[0].set_title('datos_1')
ax[1].hist(data_df['datos_2'], bins=10, color='green')
ax[1].set_title('datos_2')
ax[2].hist(data_df['datos_3'], bins=10, color='blue')
ax[2].set_title('datos_3')
plt.show()
```
